```python
imc = data["Weight"]/data["Height"]**2
data["IMC"] = imc
```

```python
data.head()
```

```python
data.drop(columns=["CALC","NCP", "SCC", "SMOKE", "family_history_with_overweight","TUE", "CAEC", "MTRANS"], inplace=True )
data.head()
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
data["Gender"] = labelencoder.fit_transform(data["Gender"])
data["NObeyesdad"] = labelencoder.fit_transform(data["NObeyesdad"])
data["FAVC"]=labelencoder.fit_transform(data["FAVC"])
data.head()
```

```python
X = data.drop("NObeyesdad", axis = 1)
y = data["NObeyesdad"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
def model_perfomance(predictions):
    print("Accuracy score on model is {}".format(accuracy_score(y_test, predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_X_train, y_train)
```

```python
log_predictions = log_model.predict(scaled_X_test)
```

```python
model_perfomance(log_predictions)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors": [3,5,7,9,11],
                "weights": ["uniform", "distance"],
              "algorithm": ["auto", "kd_tree"]}
gridkn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=2)
```

```python
gridkn.fit(scaled_X_train, y_train)
```

```python
gridkn.best_params_
```

```python
knpreds = gridkn.predict(scaled_X_test)
```

```python
model_perfomance(knpreds)
```

```python
from sklearn.svm import SVC
svm = SVC()
param_grid_svr = {"C": [0.01,0.1,0.5,1],
                   "kernel":["linear", "rbf", "poly"]}
gridsvr = GridSearchCV(svm, param_grid_svr)
```

```python
gridsvr.fit(scaled_X_train, y_train)
```

```python
pred_svc = gridsvr.predict(scaled_X_test)
model_perfomance(pred_svc)
```

```python
from sklearn.tree import DecisionTreeClassifier
param_grid = {
    "criterion": ["gini", "entropy"],
    "splitter": ["best", "random"],
    "max_depth":[None, 10,20,30,40,50],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}
grid_search_tree = GridSearchCV(estimator = DecisionTreeClassifier(), param_grid = param_grid)
```

```python
grid_search_tree.fit(scaled_X_train, y_train)
```

```python
preds = grid_search_tree.predict(scaled_X_test)
model_perfomance(preds)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
n_estimators = [32,64,128,256]
max_features = [2,3,4]
oob_score = [True, False]
```

```python
param_grid_rfc = {
    "n_estimators" : n_estimators,
    "max_features": max_features,
    'bootstrap': [True, False],
    "oob_score": oob_score
}
```

```python
grid_rfc = GridSearchCV(rfc_model, param_grid_rfc)
```

```python
grid_rfc.fit(scaled_X_train, y_train)
```

```python
preds_rfc = grid_rfc.predict(scaled_X_test)
model_perfomance(preds_rfc)
```
